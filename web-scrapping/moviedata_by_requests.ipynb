```python
import requests as req
from bs4 import BeautifulSoup as bs
import re
import pandas as pd
import numpy as np
columns_title=["year","Rank","Release","Gross","Theaters","Total Gross","Release Date","Distributor","daygrosslink"]
startyear = 2022
endyear = 1977
dfsum = pd.DataFrame()
while startyear > endyear :
    url = f"https://www.boxofficemojo.com/year/{startyear}/?grossesOption=calendarGrosses" 
    res = req.get(url)
    soup = bs(res.text,"lxml")
    elements = soup.select("table td")
    elements_movielink = soup.find_all("a", href = re.compile("^/release/"))
    text_list = [i.get_text() for i in elements if "hidden" not in i["class"]]
    movielink_list = [f"https://www.boxofficemojo.com/{i['href']}"for i in elements_movielink]
    array1 = np.array(text_list).reshape(-1,7)
    array2 = np.array(movielink_list).reshape(-1,1)
    array3 = np.full((array1.shape[0],1),startyear)
    array4 = np.concatenate((array1,array2),axis = 1)
    array4 = np.concatenate((array3,array4),axis = 1)
    dftemp = pd.DataFrame(array4,columns = columns_title)
    dfsum = dfsum.append(dftemp , ignore_index = True)
    startyear  -=  1
```

```python
# Domestic Daily by each movie  需搭配 MovieDataByYear
dfall = pd.DataFrame()
DF_Columns = ["year","title","Date","Dow","Rank","Daily","%± YD","%± LW","Theaters","Avg","To Date","Day"]
page = 100
for index , obj in enumerate(dfsum["daygrosslink"]) :
    datalist = []
    if index == page:
        break
    url = obj
    res = req.get(url) 
    soup = bs(res.text, "lxml")
    elements = soup.select("table td")
    for i in elements :
        if i.get_text() == "false" or i.get_text() == "true" :
            continue
        else:
            datalist.append(i.get_text())
    np_sum = np.array(datalist).reshape(-1,10)
    np_year = np.full((np_sum.shape[0],1),dfsum['year'][index])
    np_title = np.full((np_sum.shape[0],1),dfsum['Release'][index])
    np_sum = np.concatenate((np_title,np_sum),axis = 1)
    np_sum = np.concatenate((np_year,np_sum),axis = 1)
    dftemp = pd.DataFrame(np_sum,columns=DF_Columns)
    dfall = dfall.append(dftemp,ignore_index=True)
```
